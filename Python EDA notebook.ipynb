```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded successfully!")
```

```python
df = pd.read_excel('/content/zomato_cleaned.csv.xlsx')
df_india = df[df['Country Code'] == 1]
print("Total rows:", len(df_india))
print(df_india.head())
```

```python
top_cities = df_india['City'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='Reds_r')
plt.title('Top 10 Cities on Zomato (India)')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.tight_layout()
plt.savefig('top_cities.png')
plt.show()
print("Chart saved!")
```

```python
top_cuisines = df_india['Cuisines'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='Oranges_r')
plt.title('Top 10 Cuisines on Zomato (India)')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.tight_layout()
plt.savefig('top_cuisines.png')
plt.show()
print("Chart 2 saved!")
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Rating text', data=df_india,
              order=df_india['Rating text'].value_counts().index,
              palette='coolwarm')
plt.title('Rating Distribution of Restaurants')
plt.xlabel('Rating Category')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('rating_distribution.png')
plt.show()
print("Chart 3 saved!")
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Has Online delivery', y='Aggregate rating',
            data=df_india, palette='Set2')
plt.title('Online Delivery vs Restaurant Rating')
plt.xlabel('Has Online Delivery')
plt.ylabel('Rating')
plt.tight_layout()
plt.savefig('delivery_vs_rating.png')
plt.show()
print("Chart 4 saved!")
```
